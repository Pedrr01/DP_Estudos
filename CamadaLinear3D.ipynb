```python
import torch
#Biblioteca que fornece funções que são fundamentais para a construção e treinamento de redes neurais
from torch import nn
```

Criação de um Perceptron:

```python
#Criando um objeto do tipo nn.linear e definindo com 3 dimensões de entrada e 1 de saida:
perceptron = nn.Linear(in_features=3, out_features=1)
print(perceptron)
```

Acessando os pesos do perceptron:

```python
for nome, tensor in perceptron.named_parameters():
  print(nome, tensor.data)

#Ou acessando individualmete:
print('')
print(perceptron.weight.data)
print(perceptron.bias.data)
```

Criando uma função para construir um gráfico 3D



```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def graf3d(perceptron):
    #Convertendo os dados para array numpy no qual facilita a sua manipulação
    w1, w2, w3 = perceptron.weight.data.numpy()[0]
    b = perceptron.bias.data.numpy()

    #Gerando um array com intervalo entre -1 e 1 e distribuindo entre 10 alementos
    X1 = np.linspace(-1, 1, 10)
    X2 = np.linspace(-1, 1, 10)

    #Criando uma grade bidimensional a partir dos arrays X1 e X2
    X1, X2 = np.meshgrid(X1, X2)

    #Calcula os valores correspondentes de X3 na superfície de decisão usando a equação do plano no espaço tridimensional
    X3 = (b - w1 * X1 - w2 * X2) / w3

    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection="3d")
    ax.view_init(azim=180)

    ax.plot_surface(X1, X2, X3, cmap="plasma")

graf3d(perceptron)
```

Criando um ternsor e passando valores de entrada :

```python
x = torch.Tensor([0, -1, 2])
y = perceptron(x)

print(y)

#Gerando gráfico
graf3d(perceptron)
#Adicionando uma marcação no gráfico de acordo com as coordenadas do tensor
plt.plot([x[0]], [x[1]], x[[2]], color="r", marker="^", markersize=20)
```
