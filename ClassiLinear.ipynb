Classificação Linear

Criação de um gráfico e visualizo um ponto específico a parti dos dados:

```python
import matplotlib.pyplot as plt
#Biblioteca para gerar de conjuntos de dados artificiais para praticar exercícios:
from sklearn.datasets import make_blobs, make_classification, make_gaussian_quantiles
import numpy as np

np.random.seed(46) #Fixar o grafico da bíblioteca

#Gerando o conjunto de dados e atribuindo a X e Y
X , Y = make_classification(
    n_features=2, n_redundant=0, n_informative=1, n_clusters_per_class=1
)

#Criando do gráfico e selecionando todos os dados das linhas da coluna [0] e da coluna [1] de X
#c=Y diferencia as cores para os dados pelo rótulo Y
plt.scatter(X[:,0], X[:,1], marker="o", c=Y, edgecolor="black")

#Seleciona o 10ª valor de X
p = X[10]
#Imprime a classe(Y) associada ao 10ª valor de X
print(f"Classe do 10ª elemento de X -> {Y[10]}")
#Selecionando o 10ª elemento no gráfico a parti de suas coordenadas de X[0] e x[1]
plt.plot(p[0], p[1], marker="^", markersize=20)
```

Função para traçar a reta no gráfico:

```python
def reta(w1, w2, b):

  plt.scatter(X[:,0], X[:,1], marker="o", c=Y, edgecolor="black")
  xmin, xmax = plt.gca().get_xlim()
  ymin, ymax = plt.gca().get_ylim()

  # Valores de X vão ser fixados em um intervalo
  x = np.linspace(-2, 4, 50)

  # O valor de Y e dado pelo valor encontrado na equação onde irei isolar o Y
  # Onde w1 é o peso de X, b é o viés e w2 é o peso associado a b
  y = (-w1*x - b)/w2

  plt.axvline(0, 1, -1, color="black", linewidth=1)
  plt.axhline(0, -2, 4, color="black", linewidth=1)
  plt.plot(x,y)
  plt.grid(True)

  plt.xlim(xmin, xmax)
  plt.ylim(ymin, ymax)

#Definir valores para um modelo aleatorio
w1 = 5    #Modifica a rotação
w2 = 1    #Modifica a rotação
b = -0.4  #Desloca a reta no espaço

reta(w1, w2, b)
print(w1 * p[0] + w2 *p[1] + b)
```

Classificar os dados:

```python
def classif(ponto, w1, w2, b):
  #Calcular o resultado da equação linear para oo pontos de p
  ret = w1 * ponto[0] + w2 * ponto[1] + b

  #Verificar se o resultado é maior ou igual a zero(classe 1) ou menor que zero(classe 0)
  if ret >= 0:
    return 1, "yellow"
  else:
    return 0, "blue"

#Pontos a serem calculados
p = (2, -1)

#Chamando a função e atribuindo seu resultado as variáveis (classe, cor)
classe, cor = classif(p, w1, w2, b)
print(classe, cor)

#Chama a função da reta no gráfico e logo em seguida marca o ponto (2, -1)
reta(w1, w2, b)
plt.plot(p[0], p[1],  marker="^", color=cor, markersize=30)
```

Acurácia: métrica comum usada em problemas de classificação para avaliar o desempenho de um modelo, no qual mede a proporção de predições corretas em relação ao total de predições feitas pelo modelo.

```python
acertos = 0
for c in range(len(X)):
  #Cada valor de X vai receber classif( ) e retorna a classe 0 ou 1
  categ, _ = classif(X[c], w1, w2, b)
  #Verifica se a previsão está correta com os rótulos de Y
  if categ == Y[c]:
    acertos += 1

print("Acurácia: {0} %".format(100 * acertos/len(X)))
```
